```python
x = data['id']
y= data['age']
z= data['avg_glucose_level']
w= data['bmi']
l1 = []
l2 = []
l3 = []
l4 = []

for i in range(len(x)):
    a = l1.append(x[i])
for j in range(len(y)):
    b = l2.append(y[j])
for k in range(len(z)):
    c = l3.append(z[k])
for l in range(len(w)):
    d = l4.append(w[l])

        
print(l1)
print(l2)
print(l3)
print(l4)
```

```python
# PREDICTED VALUES BY Y = B0+B1X1+B2X2....BNXN
predicted = []

for i in range(598):
    y = 21.322987644083437 + 0.00407527 * l1[i] + 0.05609816 * l2[i] + 0.03393032 * l3[i]
#     print(y)
    predicted.append(y)
print(predicted)         
```

```python
# mean square error
# cost function = 1/2m(sum(yi(predicted)-yi(actual)))**2
b = []
for i in range(598):
    a = (predicted[i]-w[i])**2
    b.append(a)

c = sum(b)
error = (1/598)*c
print(round(error))
```

```python
##### LASSO
lasso = []
b = []
alpha = 1.0 
#lasso = MSE + alpha* sum(mod(l1[i],l2[i],l3[i]))
for i in range(5):
    a =  (l1[i]+l2[i]+l3[i])
    b.append(a)
    c = (error + alpha*sum(b))
    lasso.append(c)
print(lasso)
```

```python
# mean square error
# cost function = 1/2m(sum(yi(predicted)-yi(actual)))**2
b = []
for i in range(5):
    a = (lasso[i]-w[i])**2
    b.append(a)

c = sum(b)
error = (1/598)*c
print((error))
```

```python
##### RIDGE
RIDGE = []
b = []
alpha = 1.0 
#ridge = MSE + alpha* sum((l1[i],l2[i],l3[i]))**2
for i in range(598):
    a =  (l1[i]+l2[i]+l3[i])**2
    b.append(a)
    c = (error + alpha*sum(b))
    RIDGE.append(c)
```

```python
# mean square error
# cost function = 1/2m(sum(yi(predicted)-yi(actual)))**2
b = []
for i in range(598):
    a = (RIDGE[i]-w[i])**2
    b.append(a)

c = sum(b)
error = (1/598)*c
print((error))
```
